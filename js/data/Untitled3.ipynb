```python
import geopandas as gpd
import pandas as pd
```

```python
grid = gpd.read_file('grid_1204.geojson')
```

```python
grid .head()
```

```python
grid = grid[grid.columns.tolist()[:6]+['geometry']]
```

```python
grid.head()
```

```python
final_scores = pd.read_csv('jc_grid_Scores_final.csv')
```

```python
grid = grid.merge(final_scores,left_on = 'grid_id', right_on='grid_id')
```

```python
grid = gpd.GeoDataFrame(grid)
```

```python
grid.to_file('grid_final.geojson',driver='GeoJSON')
```
