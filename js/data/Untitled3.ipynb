```python
import pandas as pd
```

```python
tracts = pd.read_csv('jc_tract_scores_final.csv', dtype={'tract_id':str})
```

```python
tracts= tracts.iloc[:-1,:]
```

```python
tracts['tract_id'] = tracts['census_name'].str[17:]
```

```python
tracts = tracts.set_index('tract_id')
```

```python
tracts
```

```python
centerpoints = pd.read_csv('centerpoints_1210.csv')
```

```python
centerpoints = centerpoints.rename(columns={'data__name':'name','data__center_lng':'center_lng', 'data__center_lat':'center_lat'})
```

```python
centerpoints
```

```python
tracts = tracts.merge(centerpoints, left_index = True, right_on = 'name')
```

```python
tracts = tracts.set_index('name')
```

```python
tracts
```

```python
tracts.to_json('tract_info_1210.json',orient='index')
```
